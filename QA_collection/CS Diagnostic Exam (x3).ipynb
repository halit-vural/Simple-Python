## Q2:

Given N numbers and a number X, find all the occurences of X's in these numbers and replace them with X+1. X doesn't have any repeating digits. For example, if X=21 and one of the given numbers is 1221115476458721:

There are two occurences of 21 as follows: 12(21)1154764587(21)

After replacing 20's with 21's, the updated number will be: 1222115476458722

Input Format:
* Line 1: number of numbers; N <= 100
* Line 2-N+1: numbers; each number will have no more than 100 digits
* Line N+1: the number to be replaced; X < 1,000,000,000

Output Format:
* Each line will have the updated number



Example:

```
input
4
1221115476458721
32432021212121221212121212000129
21
12323122312312312321321321
21

result
1222115476458722
32432022222222222222222222000129
22
12323122312312312322322322
```

````python
inputs=['4',                                                    # 4 lines will input
        '1221115476458721', 
        '32432021212121221212121212000129', 
        '21', 
        '12323122312312312321321321', 
        '21',                                                   # num to replace
        ]

expected_outputs = [ '1222115476458722', 
                     '32432022222222222222222222000129', 
                     '22', 
                     '12323122312312312322322322', 
                     ]
````

````python
line_number = int(input())
lines = []
for i in range(line_number):
    lines.append(input())
num_to_replace = input()

#solution
for line in lines:
    new_num = int(num_to_replace) + 1
    line = line.replace(num_to_replace, str(new_num))
    print(line)
````

````python
dims = input()

rows, cols = dims.split()
rows, cols = int(rows), int(cols)

#read the first matrix
m1 = [[0 for j in range(cols)] for i in range(rows)]
for i in range(rows):
    m1[i] = input().strip().split()
    
    
input()         # empty input between matrices

#read the second matrix
m2 = [[0 for j in range(cols)] for i in range(rows)]
for i in range(rows):
    m2[i] = input().strip().split()

#sum the matrices into m3
m3 = [[int(m1[i][j])+int(m2[i][j]) for j in range(cols)] for i in range(rows)]
# m3 = [[0 for j in range(cols)] for i in range(rows)]
# for i in range(rows):
#     for j in range(cols):
#         m3[i][j] = int(m1[i][j]) + int(m2[i][j])

for i in range(rows):
    for j in range(cols):
        print(m3[i][j], end=' ')
    print()
````
