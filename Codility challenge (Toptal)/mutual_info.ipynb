```python
# Starting code

import numpy as np
import math

def mutual_information(arr):
    n = len(arr)
    m = len(arr[0])
    if n > 0 & m > 0:
        
        pi, pj = calc_p(arr)

        return calc_pij(arr) - calc_(pi) - calc_(pj)
        
    else:
        return 0.

def calc_p(arr):
    n = len(arr)
    m = len(arr[0])
    pi = np.zeros(n)
    pj = np.zeros(m)
    
    for i, row in enumerate(arr):
        for j, element in enumerate(row):
            pi[i] += element 
            pj[j] += element
    
    return pi, pj

def calc_(p):
    s = 0.
    for i, element in enumerate(p):
            s += element * np.log(element)
    return s

def calc_pij(arr):
    n = len(arr)
    m = len(arr[0])
    pij = np.zeros((n,m))

    for i, row in enumerate(arr):
        for j, element in enumerate(row):
            pij[i,j] += element * np.log(element)

    return pij
```

```python
arr = np.array([[2,3,5,4], 
                [3,7,2,1], 
                [7,3,1,2]])
```
