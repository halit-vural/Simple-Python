```python
A = [3, 1, 4, 3, 3]
X = [8, 10, 11, 13, 100]
```

```python
A = [0, 1, 0, 1, 1, 1, 0]
X = [1, 2, 3, 4, 5, 6, 7]
```

```python
# find minimum liter in cities
min_liter = max(A)
min_pos = []
min_cities = []
i = 0
for liter, city in zip(A, X):
    if liter > 0 and min_liter >= liter:
        min_liter = liter
for liter, city in zip(A, X):
    if liter == min_liter:
        min_pos.append(i)
        min_cities.append(city)
    i += 1
min_liter, min_cities, min_pos
```

```python
def check_index(list, id):
    return id >= 0 or (len(list) > id)

# find available position for starting
available_cities = []
for pos in min_pos:
    id = pos
    id_next = id + 1
    id_prev = id - 1
    if (check_index(A,id_next) and (min_liter >= (X[id_next] - X[id]))) or  (check_index(A,id_prev) and (min_liter >= (X[id] - X[id_prev]))):
        available_cities.append(X[id])
if len(available_cities):
    print(available_cities)
```

```python
def fuel_consumption(list, pos1, pos2):
    return list[pos2] - list[pos1]

def get_fuel(list, pos):
    fuel = list[pos]
    list[pos] = 0
    return fuel

# find the next city after start
curr_pos = min_pos[0]       # starting pos
count_visits = 0
cities = X.copy()
fuels = A.copy()

fuel = get_fuel(fuels, curr_pos)

def visit_city(cities, fuels, pos):
    fuel += get_fuel(fuels, pos)
    count_visits += 1
    # remove city
    fuels.pop(pos)
    cities.pop(pos)
    if check_index(cities, pos):
        if fuel >= fuel_consumption(cities, pos, pos):
            visit_city(cities, fuels, pos)
    if check_index(cities, pos-1):
        if fuel >= fuel_consumption(cities, pos-1, pos):
            visit_city(cities, fuels, pos-1)



if check_index(cities, curr_pos):
    if fuel >= fuel_consumption(cities, curr_pos, curr_pos+1):

        # travel there
        count_visits += 1
        fuel += get_fuel(fuels, curr_pos+1)
        curr_pos += 1




```
