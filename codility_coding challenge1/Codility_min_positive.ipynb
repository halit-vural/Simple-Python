find the minimum positive integer value that does not appear in the list

```python
A = [1,3,6,4,1,2]
A = [-1, -3]
A = [1, 2, 3]
A = [2,2,2]
A = [-27, -1, -27, 27]
```

```python
def find_it(A):
    max = 0     
    B = {}
    for item in A:
        if item > 0:
            B[item] = 1
            if max < item:
                max = item
    for i in range(1, max+2):
        if i not in B:
            return (i)

find_it(A)
```
