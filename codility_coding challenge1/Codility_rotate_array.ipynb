Write a function:

def solution(A, K)

that, given an array A consisting of N integers and an integer K, returns the array A rotated K times.

For example, given

    A = [3, 8, 9, 7, 6]
    K = 3
the function should return [9, 7, 6, 3, 8]. Three rotations were made:

    [3, 8, 9, 7, 6] -> [6, 3, 8, 9, 7]
    [6, 3, 8, 9, 7] -> [7, 6, 3, 8, 9]
    [7, 6, 3, 8, 9] -> [9, 7, 6, 3, 8]
For another example, given

    A = [0, 0, 0]
    K = 1

```python
A = [3, 8, 9, 7, 6]
K = 3

def solution(A, K):
    for _ in range(1):
        tmp = A[-1]
        for i in range(1,len(A)):
            A[-i] = A[-i-1]
            print(A)
        A[0] = tmp
    return A

solution(A,K)
```
