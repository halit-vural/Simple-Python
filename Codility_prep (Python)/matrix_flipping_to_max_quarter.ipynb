# Matrix Flipping
This is a mock test for flipping matrix in its rows and columns to find the maximum sum of upper left quadrant.

```python
def flippingMatrix(matrix):
    n = len(matrix)//2
    for i in range(2*n):
        invert_row(matrix,n,i)
    for i in range(n):
        invert_col(matrix, n, i)
    return sum_quadrant(matrix, n)

def invert_row(matrix, n, idx):
    sum1 = sum2 = 0
    for i in range(n):
        sum1 += matrix[idx][i]
        sum2 += matrix[idx][-i-1]
    if sum1 < sum2:
        #invert
        print('inverting row:', idx)
        for i in range(n):
            matrix[idx][i], matrix[idx][-i-1] = matrix[idx][-i-1], matrix[idx][i]
        print(matrix)
        
    # return matrix
    

def invert_col(matrix, n, idx):
    sum1 = sum2 = 0
    for i in range(n):
        sum1 += matrix[i][idx]
        sum2 += matrix[-i-1][idx]
    if sum1 < sum2:
        #invert
        print('inverting col:', idx)
        for i in range(n):
            matrix[i][idx], matrix[-i-1][idx] = matrix[-i-1][idx], matrix[i][idx]
        print(matrix)
    # return matrix

def sum_quadrant(matrix, n):
    sum_q = 0
    for i in range(n):
        for j in range(n):
            sum_q += matrix[i][j]
    return sum_q
```
