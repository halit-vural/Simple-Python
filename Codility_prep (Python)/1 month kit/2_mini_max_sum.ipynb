```python
ex1 = [1, 3, 5, 7, 9]
ex2 = [14, 3, 9, 0, 4, 1]
ex3 = [11,1,3,5,7,2,9]          # --> 27 37

arr = ex1

def miniMaxSum(arr):
    arr.sort()

    min_tot = max_tot = total = 0 
    for i in arr:
       total += i
    min_tot = total - arr[-1]
    max_tot = total - arr[0]

    print(min_tot, max_tot)


miniMaxSum(arr)
```
