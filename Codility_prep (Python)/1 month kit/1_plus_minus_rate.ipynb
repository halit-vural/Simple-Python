Given an array of integers, calculate the ratios of its elements that are positive, negative, and zero. Print the decimal value of each fraction on a new line with 6 places after the decimal.

```python
ex1 = [1,1,0,-1,-1]
ex2 = [-4, 3, -9, 0, 4, 1]

arr = ex2


def plusMinus(arr):
    pos_tot = neg_tot = zero_tot = 0
    n = len(arr)
    if n <= 0:
        return
    for i in arr:
        if i < 0:
            neg_tot += 1
        elif i == 0:
            zero_tot += 1
        elif i > 0:
            pos_tot += 1
    pos_rate = pos_tot / n
    neg_rate = neg_tot / n
    zero_rate = zero_tot / n
    print(pos_rate)
    print(neg_rate)
    print(zero_rate)


plusMinus(arr)
```
