```python
def permute(arr):
    if len(arr) == 1:
        yield arr
    else:
        for i in range(len(arr)):
            for perm in permute(arr[:i] + arr[i+1:]):
                yield [arr[i]] + perm

for i in permute(C):
    print(i)
```

```python
import time
import functools
def stop_watch(func):
  @functools.wraps(func)
  def wrapper(*args, **kwargs):
    start = time.perf_counter() # Start Time
    result = func(*args, **kwargs)
    end = time.perf_counter() # End Time
    print(f"{func.__name__} ran in {end - start:0.4f} seconds")
    return result
  return wrapper

@stop_watch
def some_function(sec: int):
  time.sleep(sec)

if __name__ == "__main__":
  some_function(5)
```
